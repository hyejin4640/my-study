```python
# 메뉴 프로그램
from tkinter import *
import random

win = Tk() 
win.geometry("390x180")

menuselect=0

display1 = IntVar()
display2 = IntVar()
display3 = IntVar()
display4 = IntVar()

def cost():
    total = 6000 * int(display1.get()) + 7000 * int(display2.get()) + 8000 * int(display3.get()) + 9000 * int(display4.get()) 
    lblmoney2.config(text=total)

def menu1():
    global menuselect
    lblmenu1.config(text = "짜장면")
    menuselect = 1
def menu2():
    global menuselect
    lblmenu1.config(text = "간짜장")
    menuselect = 2
def menu3():
    global menuselect
    lblmenu1.config(text = "짬뽕")
    menuselect = 3
def menu4():
    global menuselect
    lblmenu1.config(text = "볶음밥")
    menuselect = 4
def up():
    if menuselect == 1:
        key = int(display1.get()) + 1
        display1.set(key)
    if menuselect == 2:
        key = int(display2.get()) + 1
        display2.set(key)
    if menuselect == 3:
        key = int(display3.get()) + 1
        display3.set(key)
    if menuselect == 4:
        key = int(display4.get()) + 1
        display4.set(key)
    cost()
    
def down():
    if menuselect == 1:
        key = int(display1.get()) - 1
        display1.set(key)
    if menuselect == 2:
        key = int(display2.get()) - 1
        display2.set(key)
    if menuselect == 3:
        key = int(display3.get()) - 1
        display3.set(key)
    if menuselect == 4:
        key = int(display4.get()) - 1
        display4.set(key)
    cost()
    


btnA0 = Button(win, width=9, height=4, text="짜장면", command=menu1); btnA0.pack()
lblA0 = Label(win, width=5, height=2, text = "6000", relief="ridge"); lblA0.pack()
lblA1 = Label(win, width=5, height=2, text = "0", textvariable=display1, relief="ridge"); lblA1.pack()

btnB0 = Button(win, width=9, height=4, text="간짜장", command=menu2); btnB0.pack()
lblB0 = Label(win, width=5, height=2, text = "7000", relief="ridge"); lblB0.pack()
lblB1 = Label(win, width=5, height=2, text = "0", textvariable=display2, relief="ridge"); lblB1.pack()

btnC0 = Button(win, width=9, height=4, text="짬뽕", command=menu3); btnB0.pack()
lblC0 = Label(win, width=5, height=2, text = "8000", relief="ridge"); lblB0.pack()
lblC1 = Label(win, width=5, height=2, text = "0", textvariable=display3, relief="ridge"); lblB1.pack()

btnD0 = Button(win, width=9, height=4, text="볶음밥", command=menu4); btnB0.pack()
lblD0 = Label(win, width=5, height=2, text = "9000", relief="ridge"); lblB0.pack()
lblD1 = Label(win, width=5, height=2, text = "0", textvariable=display4, relief="ridge"); lblB1.pack()

lblmenu0 = Label(win, width=11, height=1, text="품목선택")
lblmenu1 = Label(win, width=11, height=1, text="", relief="ridge")

lblcnt0 = Label(win, width=11, height=1, text="수량")
btncnt1 = Button(win, width=4, height=1, text="+", command=up)
btncnt2 = Button(win, width=4, height=1, text="-", command=down)

lblmoney1 = Label(win, width=11, height=1, text="총금액")
lblmoney2 = Label(win, width=11, height=1, text="", relief="ridge")

btnA0.place(x=10, y=10); lblA0.place(x=85, y=10); lblA1.place(x=85, y=45)
btnB0.place(x=140, y=10); lblB0.place(x=215, y=10); lblB1.place(x=215, y=45)
btnC0.place(x=10, y=95); lblC0.place(x=85, y=95); lblC1.place(x=85, y=130)
btnD0.place(x=140, y=95); lblD0.place(x=215, y=95); lblD1.place(x=215, y=130)

lblmenu0.place(x=290, y=10); lblmenu1.place(x=290, y=30)
lblcnt0.place(x=290, y=60); btncnt1.place(x=290, y=82); btncnt2.place(x=335, y=82)
lblmoney1.place(x=290, y=122)
lblmoney2.place(x=290, y=145)

win.mainloop()
```
