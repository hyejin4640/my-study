```python
from tkinter import *
from functools import partial

win = Tk()
win.geometry("510x250")
# win.config(background=)

  #for나 함수밖에서도 변수를 사용하려면 global변수를 사용해야하는데 , 함수내에서 값이 변하면 global을 붙여 사용해야한다
display = StringVar()

turn = False
parkNo =[]

for i in range(8):
    parkNo.append(StringVar())

def carIn(carnumNow):
    for i in  range (len(parkNo)):
        if parkNo[i].get() == "":
            parkNo[i].set(carnumNow)
            return
        else:
            continue
    
def showNum(now):
    global turn
    if turn :
        display.set(""+now)
        turn = False
    else :
        n = display.get()
        n+=now
        display.set(n)
    
def clickcarIn():
    global turn
    carnumNow = display.get()
    if len(carnumNow) != 4 :
        display.set("차량 뒷번호 4자리를 입력하세요")
        turn = True
    else :
        carIn(carnumNow)
        display.set("")
        
def clickCarOut():
    global turn
    carnumNow = display.get()
    for i in range(len(parkNo)):
        if parkNo[i].get() == carnumNow:
            parkNo[i].set("")
            turn=True
            break
    if i == len(parkNo):
        display.set("일치하는 차량번호가 없습니다")
        turn=True
        
#         else:
#             display.set("일치하는 차량번호가 없습니다")
#             turn=True


            
        
# ul ---------------------------------------------------------------------------------------------
#주차 타이틀창
lblTitle = Label(win,relief="ridge",text="주차상태")
lblTitle.place(x=10,y=10,width=320,height=40)

#주차위치 번호창

lblNoList=[]
for i in range(8):
    lblNoList.append(Label(win,relief="ridge",text=str(i+1)))
    lblNoList[i].config(background="lightblue")
for i in range(4):
    lblNoList[i].place(x=10 ,y=(60+45*i),width=40,height=40)
    lblNoList[i+4].place(x=180 , y=(60+45*i), width=40 , height=40)

# 주차 상태창
lblCarList =[]
for i in range(8):
    lblCarList.append(Label(win, relief="ridge",textvariable=parkNo[i]))  
for i in range(4):
    lblCarList[i].place(x=60 , y=(60+45*i),width=100,height=40)
    lblCarList[i+4].place(x=230 , y=(60+45*i), width=100 , height=40)

#번호 출력창
lblCarNo = Label(win,relief="ridge",textvariable=display)
lblCarNo.place(width=140,height=40,x=360,y=10)


#번호 입력창

btnNoList = []
for i in range(9):
    btnNoList.append(Button(win, text=str(i+1) ,command=partial(showNum , str(i+1))))
    
for i in range(3):
    btnNoList[i].place(x=360+(50*i),y=60,width=40,height=40)
    btnNoList[i+3].place(x=360+(50*i),y=105,width=40,height=40)
    btnNoList[i+6].place(x=360+(50*i),y=150,width=40,height=40)
    
    
# IN , OUT , 0 버튼 따로 생성
btnIn=Button(win,text="IN" , command=clickcarIn)
btnIn.place(x=360,y=195,width=40,height=40)
btnIn.config(background="lightblue")

btnOut=Button(win,text="OUT",command=clickCarOut)
btnOut.place(x=460,y=195,width=40,height=40)
btnOut.config(background="lightblue")

btnNoList.append(Button(win,text="0", command=partial(showNum , str(0))))
btnNoList[9].place(x=410,y=195,width=40,height=40)





    
win.mainloop()
```
